```python
root_source  =  "DATA_CHUP\DATA_CHUP_test"
root_target = "DATA_CHUP\\DATA_CHUP_test\\DATA_CHUP_test_labels"
if not os.path.exists(root_target):
    os.makedirs(root_target)

image_folder_name_list = os.listdir(os.path.join(root_source, "DATA_CHUP_test_images"))

for image_folder_name in image_folder_name_list:
    image_folder_name_path = os.path.join(root_source, "DATA_CHUP_test_images", image_folder_name)    
    image_folder_name_path = os.path.join(root_source, "DATA_CHUP_test_images", image_folder_name)
    image_file_name_list = os.listdir(image_folder_name_path)
    for image_file_name in image_file_name_list:
        label_file_name = image_file_name.replace(".jpg", ".txt")
        label_file_name_path = os.path.join(root_source, "DATA_CHUP_test_labels", label_file_name)
        with open(label_file_name_path, "w") as f:
            f.write(image_folder_name.split("_")[0])
            f.close()
            print(image_folder_name.split("_")[0])
```

# Chuyển ảnh 

```python
root_source  =  "DATA_CHUP\DATA_CHUP_test\\DATA_CHUP_test_images"
root_target = "DATA_CHUP\\DATA_CHUP_yolo\\images"
```

```python
import os 
import shutil
image_folder_list = os.listdir(root_source)
for image_folder_name in image_folder_list:
    image_folder_name_path = os.path.join(root_source, image_folder_name)
    image_file_name_list = os.listdir(image_folder_name_path)
    for image_file_name in image_file_name_list:
        image_file_name_path = os.path.join(image_folder_name_path, image_file_name)
        image_file_name_path_target = os.path.join(root_target, image_file_name)
        shutil.copyfile(image_file_name_path, image_file_name_path_target)
```
