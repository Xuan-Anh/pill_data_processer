```python
target_path = "data_video_sau\\v9\\test_rename\\labels"
predict_path = "data_video_sau\\v9\\labels_rename_result"
```

```python
import os
predict_file = os.listdir(predict_path)
target_file = os.listdir(target_path)
```

```python
predict_list = []
target_list = []
for file in predict_file:
    with open(os.path.join(predict_path, file), 'r') as f:
        line = f.readline()
        try:
            predict_list.append(int(line.split()[0]))
        except:
            print("error in predict")

for file in target_file:
    with open(os.path.join(target_path, file), 'r') as f:
        line = f.readline()
        try:
            target_list.append(int(line.split()[0]))
        except:
            print("error in target")
            print(file)

            

```
