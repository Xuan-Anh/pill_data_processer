```python
root_data_v10 =  "DATA_XA_sorted\\v10"
root_data_v10_5shots =  "DATA_XA_sorted\\v10_5shots"
```

```python
import random
import os
import shutil
folder_name_list = os.listdir("{}/images".format(root_data_v10))
for index, folder_name in enumerate(folder_name_list):
    folder_images_path = "{}/images/{}".format(root_data_v10, folder_name)
    file_image_name_list = os.listdir(folder_images_path)
    print("folder_name: {}".format(folder_name))

    # print("AAAAAAAAAAAAAAAAAAAAAAA")

    random_list = []

    if len(file_image_name_list) > 15:
        random_list = random.sample(file_image_name_list, 15)
    else:
        random_list = file_image_name_list


    for file_image_name in random_list:

        source_file_image_path = "{}/{}".format(folder_images_path, file_image_name)
        source_file_label_path = "{}/labels/{}/{}".format(root_data_v10, folder_name, file_image_name.replace(".jpg", ".txt"))
        
        target_image_folder_path = "{}/images/{}".format(root_data_v10_5shots, folder_name)
        target_label_folder_path = "{}/labels/{}".format(root_data_v10_5shots, folder_name)
        if not os.path.exists(target_image_folder_path):
            os.makedirs(target_image_folder_path)
        if not os.path.exists(target_label_folder_path):
            os.makedirs(target_label_folder_path)

        target_image_file_path = "{}/images/{}/{}".format(root_data_v10_5shots, folder_name, file_image_name)
        target_label_file_path = "{}/labels/{}/{}".format(root_data_v10_5shots, folder_name, file_image_name.replace(".jpg", ".txt"))

        print("source_file_image_path: {}".format(source_file_image_path))
        print("source_file_label_path: {}".format(source_file_label_path))
        print("target_file_image_path: {}".format(target_image_file_path))
        print("target_file_label_path: {}".format(target_label_file_path))

        shutil.copyfile(source_file_image_path, target_image_file_path)
        shutil.copyfile(source_file_label_path, target_label_file_path)







```
